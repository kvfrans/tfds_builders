```python
import tensorflow_datasets as tfds
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = tfds.load('imagenet2012', split='train')
def deserialization_fn(data):
    image = data['image']
    min_side = tf.minimum(tf.shape(image)[0], tf.shape(image)[1])
    image = tf.image.resize_with_crop_or_pad(image, min_side, min_side)
    image = tf.image.resize(image, (256, 256), antialias=True)
    return image, data['label']
dataset = dataset.map(deserialization_fn, num_parallel_calls=tf.data.AUTOTUNE)
dataset = dataset.batch(8)
dataset = tfds.as_numpy(dataset)
dataset = iter(dataset)

image, label = next(dataset)
print(image.shape, np.max(image), np.min(image)) # Between 0 and 256.
print(label.shape)

# plot in a row.
fig, axs = plt.subplots(1, 8, figsize=(20, 20))
for i in range(8):
    axs[i].imshow(image[i] / 256)
    axs[i].axis('off')
plt.show()
```
